# Feature Engineering

### Drop columns that will not be used

```python
df_cases = df_cases.drop(columns=['UID', 'iso2', 'iso3', 'code3', 'Country_Region', 'Lat', 'Long_', 'Combined_Key'])
df_cases
```

### Unpivot a DataFrame from wide to long format, rename columns

```python
df_cases = df_cases.melt(id_vars=df_cases.columns[0:3])
df_cases = df_cases.rename(columns={"variable": "date", "value": "cum_case", 
                               "Province_State": "state", "Admin2": "county",
                              "FIPS": "COUNTYFP"})
df_cases
```

### Drop rows with missing values

```python
df_cases = df_cases.dropna()
```

### Changed dataframe types

```python
df_cases.info()
```

```python
df_cases['date']= pd.to_datetime(df_cases['date'])
df_cases['COUNTYFP'] = df_cases['COUNTYFP'].astype(np.int64)
```

### Sort Dataframe 

```python
df_cases = df_cases.sort_values(by=['state', 'county', 'date'], ascending = [True, True, True]).reset_index(drop=True)
df_cases
```

### Join cases dataframe with mask use dataframe

```python
df_mask_use
```

```python
df_cases = pd.merge(df_cases, df_mask_use, on='COUNTYFP', how='inner')
```

```python
df_cases
```

### Feature engineering for mandates dataframe

```python
# lets set N/A values from before mandates came into effect to no. 
# Technically people were not required to wear masks when covid did not exist in the state, no works here.
df_mandates["Face_Masks_Required_in_Public"] = df_mandates["Face_Masks_Required_in_Public"].fillna("No")
df_mandates = df_mandates.drop(columns=['order_code', 'Source_of_Action', 'URL', 
                                        'Citation', 'State_Tribe_Territory', 'County_Name'])
df_mandates
```

```python
df_mandates.info()
```

```python
df_mandates['date']= pd.to_datetime(df_mandates['date'])
df_mandates['FIPS_State'] = df_mandates['FIPS_State'].astype(str)
df_mandates['FIPS_County'] = df_mandates['FIPS_County'].astype(str)
df_mandates['COUNTYFP'] = (df_mandates['FIPS_State'] + df_mandates['FIPS_County'])

df_mandates = df_mandates.drop(columns=['FIPS_State', 'FIPS_County'])

df_mandates['COUNTYFP'] = df_mandates['COUNTYFP'].astype(np.int64)
```

```python
df_mandates['Face_Masks_Required_in_Public'] = pd.Series(np.where(df_mandates.Face_Masks_Required_in_Public.values 
                                                                  == 'Yes', 1, 0), df_mandates.index)
```

```python
df_mandates
```

### Merge mandas dataframe with cases dataframe
Do a left join with df_cases and where Face_Masks_Required_in_Public is not specified change that value to zero.

```python
df_cases = pd.merge(df_cases, df_mandates, on=['COUNTYFP', 'date'], how='left')
df_cases = df_cases.drop_duplicates()
```

```python
df_cases['Face_Masks_Required_in_Public'] = df_cases['Face_Masks_Required_in_Public'].fillna(0)
df_cases
```

### Drop rows before 2020-04-15 and two years after which is 2022-04-15

```python
df_cases = df_cases[(df_cases['date'] >= '2020-04-15') & (df_cases['date'] < '2022-04-15')]
df_cases
```

### Drop the few counties that have a different length time series to Wayne, Michigan
These counties have incorrect data as there is 730 days between 2020-04-15 and 2022-04-15

```python
unique_state_county_df = df_cases[['state', 'county']].drop_duplicates()
to_drop = []
for i in range(unique_state_county_df.shape[0]):
    county_df = df_cases[df_cases['state'] == unique_state_county_df.iloc[i]['state']]
    state_county_df = county_df[county_df['county'] == unique_state_county_df.iloc[i]['county']]
    
    if state_county_df.shape[0] != 730:
        to_drop.append([unique_state_county_df.iloc[i]['state'], unique_state_county_df.iloc[i]['county']])
    # print(unique_state_county_df.iloc[i]['state'])
    # print(unique_state_county_df.iloc[i]['county'])
    # print('count:', state_county_df.shape[0])
```

```python
to_drop
```

```python
for (state, county) in to_drop:
    df_cases = df_cases.drop(df_cases[(df_cases['state'] == state) & (df_cases['county'] == county)].index)
```

### Save where case count when county change happens. This is where the next line of code fails

```python
cases_to_fill = df_cases.loc[~(df_cases['county'] == df_cases.shift(periods=1)['county'])]['cum_case']
cases_to_fill
```

### Change from cumulative cases to new cases for each day

```python
df_cases['cum_case'] = df_cases['cum_case'] - df_cases.shift(periods=1)['cum_case']
```

```python
for index in cases_to_fill.index:
    df_cases.at[index, 'cum_case'] = cases_to_fill[index]
```

```python
df_cases = df_cases.rename(columns={"cum_case":  "new_cases"})
```

```python
df_cases
```

### Windowing data
Case numbers are not always reported for an indidual day and that count is added to subsqeuent days. I will smooth the graph by taking the average of a three day window for each day

```python
df_cases['new_cases'] = df_cases['new_cases'].rolling(2).mean()
```

```python
for index in cases_to_fill.index:
    df_cases.at[index, 'new_cases'] = cases_to_fill[index]
```

```python
df_cases
```

# Change mask usage columns to one column called "mask_use"
The columns "NEVER", "RARELY", "SOMETIMES", "FREQUENTLY", and "ALWAYS" can be combined by multiplying the percentage represented in each column by 0 for "NEVER, 0.25 for "RARELY", 0.5 for "SOMETIMES", 0.75 for "FREQUENTLY", and  1 for "ALWAYS". Then adding these columns together will give us a nominal continuous representation for mask usage between 0 and 1. After, I set Face_Masks_Required_in_Public to NA for all days afte 2021-04-15 where the survey data begins to become invalid due to data drift.

```python
df_cases["mask_use"] = (df_cases['RARELY'] * 0.25 + df_cases['SOMETIMES'] * 0.5 + 
                        df_cases['FREQUENTLY'] * 0.75 +  df_cases['ALWAYS'])
```

```python
df_cases.loc[df_cases['date'] >= '2021-04-15', 'mask_use'] = np.nan
```

```python
df_cases = df_cases.drop(columns=["NEVER", "RARELY", "SOMETIMES", "FREQUENTLY", "ALWAYS"])
df_cases
```
